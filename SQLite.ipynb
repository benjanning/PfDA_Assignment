## SQLite in Python

```python
import sqlite3
```

```python
# Create tutorial file

con = sqlite3.connect("data/tutorial.db")
```

```python
# Cursor
# an object that is used to make the connection for executing SQL queries. 
# It acts as middleware between SQLite database connection and SQL query. 
# It is created after giving connection to SQLite database.

cur = con.cursor()
```

```python
# Create a Table. Specify that name and address are text.

cur.execute("CREATE TABLE person(name text, address text)")
con.commit()
```

```python
# Select table from a Database. Master maintains track of all the other tables.

cur.execute("SELECT name FROM sqlite_master WHERE type='table'")
cur.fetchall()
```

```python
# Insert name and address into the Person Table

cur.execute("INSERT INTO person VALUES ('Kathleen Mayer','Westmeath')")
       
```

```python
# Select everything from Person

cur.execute("SELECT * FROM Person")
cur.fetchall()
```

```python
# Delete from Person

cur.execute("DELETE FROM Person")
con.commit()
```

```python
# Insert More Rows

people = [
    ("Ashley Adams", "Leitrim",),
    ("Brenda Snyder", "Dublin",),
    ("Brenda Morrison", "Kildare",),
    
]
cur.executemany("INSERT INTO person VALUES(?, ?)", people)
con.commit()
```

```python
# Select all from People

cur.execute("SELECT * FROM person")
cur.fetchall()
```

```python
# Select hidden ROWID and all names and addresses from Person
# Auto Incremenets - unique number to be generated automatically when a new record is inserted into a table

cur.execute("SELECT ROWID, name, address * FROM Person")
cur.fetchall()
```

```python
cur.execute("DELETE FROM person WHERE ROWID=1")
con.commit()
```

```python
# Do Selects

cur.execute("SELECT ROWID, name, address * FROM Person")
cur.fetchall()
```

```python
# Adding Kathleen back into the Databse under a new ROWID

cur.execute("INSERT INTO person VALUES ('Kathleen Mayer','Westmeath')")
con.commit()
```

```python
# Show this.
# Technically now there is no duplicate rows (if people have the same name in rows, they have different numbers now)
# These keys are called Primary Keys

cur.execute("SELECT ROWID, name, address * FROM Person")
cur.fetchall()
```
